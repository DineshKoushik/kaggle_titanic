```python
# import pandas
import pandas as pd

# read train csv file
df_train = pd.read_csv("train.csv")

# read test csv file
df_test = pd.read_csv("test.csv")
```

```python
df_train.drop(['PassengerId','Name', 'Ticket', 'Cabin'],axis = 1, inplace = True)
df_train.head(10)
```

```python
PassengerId_test = df_test['PassengerId']
df_test.drop(['PassengerId','Name', 'Ticket', 'Cabin'],axis = 1, inplace = True)
df_test.head(10)
```

```python
# one-hot encoding
df_train = pd.get_dummies(df_train)
df_train.head(10)
```

Checking for null values in the train set

Assigning the mean values to the null values

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
```

verifying for null values in the train set

```python
# one-hot encoding
df_test = pd.get_dummies(df_test)
df_test.head(10)
```

Checking for null values in the test set

```python
df_test.isnull().sum().sort_values(ascending=False).head(10)
```

Assigning the mean values to the null values

```python
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
df_test['Fare'] = df_test['Fare'].fillna(df_test['Fare'].mean())
```

Verifying for null values in the test set

**Features**

```python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']
```

**Defining features and prediction target**

```python
X = df_train[features]
y = df_train.Survived
```

```python
# import train_test_split
from sklearn.model_selection import train_test_split


# Split into training and validation data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0, test_size=0.3)
```

```python
# import logistic regression
from sklearn.linear_model import LogisticRegression
# import MAE
from sklearn.metrics import mean_absolute_error

lr = LogisticRegression(max_iter=2000)

# fit the model
lr.fit(train_X, train_y)

val_predictions = lr.predict(val_X)
val_mae1 = mean_absolute_error(val_y, val_predictions)
val_mae1
```

```python
from sklearn.tree import DecisionTreeRegressor

# define regressor with no max_leaf_nodes
regressor = DecisionTreeRegressor(random_state=0)

# fit the model
regressor.fit(train_X, train_y)

val_predictions = regressor.predict(val_X)
val_mae2 = mean_absolute_error(val_y, val_predictions)
val_mae2
```

```python
# define regressor with 50 max_leaf_nodes
regressor_l = DecisionTreeRegressor(max_leaf_nodes=50, random_state=0)

# fit the model
regressor_l.fit(train_X, train_y)
val_predictions = regressor_l.predict(val_X)
val_mae3 = mean_absolute_error(val_y, val_predictions)
val_mae3
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(random_state = 0)

# fit the model
rfr.fit(train_X, train_y)
val_predictions = rfr.predict(val_X)
val_mae4 = mean_absolute_error(val_y, val_predictions)
val_mae4
```

```python
# create X_test from the df_test(from the above)
X_test = df_test[features]
```

```python
# make predictions for the competition
test_pred = rfr.predict(X_test)
print(test_pred)
```

```python
# Save test predictions to file
output = pd.DataFrame({'PassengerID': PassengerId_test,
                       'Survived': test_pred})
output.to_csv('submission.csv', index=False)
```
