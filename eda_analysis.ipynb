**Features**

```python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']
```

**Defining features and prediction target**

```python
X = df_train[features]
y = df_train.Survived
```

```python
# import train_test_split
from sklearn.model_selection import train_test_split


# Split into training and validation data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0, test_size=0.3)
```

```python
# import logistic regression
from sklearn.linear_model import LogisticRegression

# import accuracy_score
from sklearn.metrics import  accuracy_score 

lr = LogisticRegression(max_iter=2000)

# fit the model
lr.fit(train_X, train_y)

val_predictions = lr.predict(val_X)
accuracy_score(val_y, val_predictions)
```

```python
# import logistic regression
from sklearn.ensemble import RandomForestClassifier

# import accuracy_score
from sklearn.metrics import  accuracy_score 

rfc = RandomForestClassifier(n_estimators=700, random_state=1)

# fit the model
rfc.fit(train_X, train_y)

val_predictions = rfc.predict(val_X)
accuracy_score(val_y, val_predictions)
```

```python
# create X_test from the df_test(from the above)
X_test = df_test[features]
```

```python
# make predictions for the competition
test_pred = rfc.predict(X_test)
print(test_pred)
```

```python
# Save test predictions to file
output = pd.DataFrame({'PassengerID': PassengerId_test,
                       'Survived': test_pred})
output.to_csv('submission.csv', index=False)
```
